### The Products Table

Assuming that each item has one product-code, I create a new table that summarizes the data for each item as follows:

```python
# calculate table fields
products = desc2.merge(df.groupby('StockCode').agg(
    {'UnitPrice':'mean','CustomerID':'nunique', 'Quantity':'sum',
     'TotalBill':'sum'}).sort_values(by='UnitPrice', ascending=False).reset_index(), how='outer',
                       on='StockCode')

# calculate products periods
products_period = df.groupby('StockCode').InvoiceDate.min().reset_index().merge(
    df.groupby('StockCode').InvoiceDate.max(),on='StockCode').rename(
    columns={'InvoiceDate_x':'first_date','InvoiceDate_y':'last_date'})
products_period['period'] = (products_period['last_date'] - products_period['first_date'] +
                             timedelta(days=1)).dt.days

# calculate unique sales
one_item_invoices = df.groupby('InvoiceNo')['StockCode'].nunique().reset_index().query(
    'StockCode==1')['InvoiceNo'].unique().tolist()
nunique_sales = df.query('InvoiceNo in @one_item_invoices')[['InvoiceNo','StockCode']].groupby(
    'StockCode')['InvoiceNo'].nunique().reset_index()

# marging the calculation
products = products.merge(df[df['Quantity']>0].groupby('StockCode').agg(
    {'InvoiceNo':'nunique','Quantity':'sum'}).reset_index().rename(columns={'Quantity':'Sold'}),
                          how='outer', on='StockCode').fillna(0)
products = products.merge(df[df['Quantity']<0].groupby('StockCode')['Quantity'].sum().abs().reset_index().rename(
    columns={'Quantity':'%Returns'}), how='outer', on='StockCode').fillna(0)
products['%Returns'] = products['%Returns'] / products['Sold']
products = products.merge(products_period.drop('last_date',axis=1), how='outer', on='StockCode')
products = products.merge(nunique_sales, how='outer', on='StockCode').fillna(0)
products['%UnSa'] = (products['InvoiceNo_y'] / products['Sold'])

# columns renaming
products.rename(columns={'InvoiceNo_x':'Invoices','CustomerID':'Clients','InvoiceNo_y':'UnSa','TotalBill':'TotalRevenue'}, inplace = True)

print('>> \033[1mProducts Table\033[0m:\n')
print('\tNumber of entries:\t{:,d}'.format(products.shape[0]))
display(products.head())
```

The products table summaize the data about 3,957 items for each `StockCode` as follows:
* `Description2` - the longest Descriptions
* `UnitPrice` -   average price per unit
* `Clients` -     number of unique registered clients
* `Quantity` -     total number of units in data (sold + returned)
* `TotalRevenue`
* `Sold` -        total amount of units sold
* `Invoices` -    total amount of invoices (no returns included)
* `%Returns` -    % of negtive total quantities out from total amount sold
* `first_date` -  first apeareance of product in the log
* `period` -      total days the product was sold
* `UnSa` -      stands for Unique Sales, number of times the product was the only item on the invoice
* `%UnSa` -      % of times the product was the only item on the invoice from total number of invoices

The following function will help me to clear the data by a specific query:

```python
def pclean(df ,query_str):
    len0 = df.shape[0]
    new_tbl = df.query(query_str)
    len1 = new_tbl.shape[0]
    lost = len0 - len1
    print('{:,d} entries ({:.3%} of total) been filtred out from the products table.'.format(lost,lost/len0))
    print('New number of products:\t {:,d}'.format(len1))
    return new_tbl
```

#### Outliers Detection

It would be wise to continue searching for non-product items by searching for outliers in the new products table. The method I will use for outliers detection is to set a lower limit as the 5th percentile and an upper limit as the 95th percentile:

```python
def outliers(col):
    uplim = np.percentile(col,95)
    lolim = np.percentile(col,5)
    above = (col > uplim).sum()
    below = (col < lolim).sum()
    total = len(col)
    print('Outliers in \033[1m{:}\033[0m\n'.format(col.name))
    print('Min value:\t {: >10.2f} |\nMax value:\t {: >10.2f} |'.format(col.min(),col.max()))
    print('\n5th percentile:\t {: >10.2f} |\n95th percentile: {: >10.2f} |'.format(lolim,uplim))
    print('\nEntries below:\t {: >10,d} | {:.2%} of total\nEntries above:\t {: >10,d} | {:.2%} of total\n'.format(
        below,below/total,above,above/total))
```

## Revenues by colors

During the time I have examined the variety of products in this paper, I have noticed a considerable amount of products whose description contains information about their color. After discovering that products purchased together the most times are usually the same product of a different sample or in different colors, I decided to examine which colors are the most profitable for the store. At this point I am interested in checking if there is a significant difference in profits between colors.

First I have to extract the color name from the product description.

### Colors extraction from Description

```python
# !pip install webcolors
import webcolors
colors = list(webcolors.CSS3_NAMES_TO_HEX)
```

```python
from nltk.stem import SnowballStemmer
engs = SnowballStemmer('english')
```

```python
#nltk.download('stopwords')
from nltk.corpus import stopwords
stps = set(stopwords.words('english'))
```

```python
def detect_color(row):
    desc = row['Description2']
    key_items = []
    for word in desc.split(' '):
        stemw = engs.stem(word)
        if stemw in colors:
            key_items.append(stemw)
    return ', '.join(key_items)
```

```python
products['Colors'] = products.apply(detect_color, axis=1).replace('','none')
products[['Description2','Colors']].head()
```

I will now check how many products I have from any color or combination of colors:

```python
products.groupby('Colors')['StockCode'].nunique().sort_values(ascending=False).reset_index().rename(
    columns={'StockCode':'Products'})
```

The color combinations exist in too small quantities, also it is not possible to know which color is more dominant. Therfore from the list above I will choose only the singular colors that exist in at least 10 products.

```python
color_quan = products.groupby('Colors')['StockCode'].nunique().sort_values(ascending=False).reset_index()
top_colors = color_quan.query('Colors !="none" & StockCode>10')['Colors'].tolist()
top_colors
```

### Comparing the average revenue from each color

```python
color_avg_rev = products.query('Colors in @top_colors').groupby(
    'Colors')['TotalRevenue'].mean().sort_values(ascending=False).reset_index()
height = color_avg_rev.TotalRevenue
bars = color_avg_rev.Colors
x_pos = np.arange(len(bars))
```

```python
plt.figure(figsize=(15,5))
ax = plt.bar(x_pos, height, color=bars)
ax2 = sns.scatterplot(x=color_avg_rev.Colors, y=color_avg_rev.TotalRevenue, marker=False)
for dot in ax2.patches:
    ax2.annotate(format(dot.get_height(), ',.2f'),
                   (dot.get_x() + dot.get_width() / 2,
                    dot.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.xticks(x_pos, bars)
plt.title('Average Revenue per Product-color')
plt.show()
```

```python
color_prod = products.query('Colors in @top_colors').groupby('Colors')['StockCode'].nunique().sum()
total_prod = products['StockCode'].nunique()
print('Total amount of products:\t\t{:,d}'.format(total_prod))
print('Total amount of products with color:\t{:,d} ({:.1%})'.format(color_prod,color_prod/total_prod))
```

It seems that on average the color red is more profitable by far from the rest of the colors, later leads the colors white, yellow, silver and pink. But this is only a very small sample, after all, every product in the store has at least one dominant color, and the data I have represent only 28% of all products.

In order to know that indeed the difference between the averages is significant enough to contain the conclusions on all of the products in the store, I will formulate and test some hypotheses. Yet again I will write a function to help me run multiple tests:

```python
from scipy import stats as st
```

```python
def rev_ttest(c1,c2,alpha=.05):
    sample_1 = products.query('Colors == @c1')['TotalRevenue']
    sample_2 = products.query('Colors == @c2')['TotalRevenue']
    results = st.ttest_ind(sample_1, sample_2)
    print("""
Null hypothesis:
    There is no significant difference between the average revenue from products
        with the color \033[1m{:}\033[0m and products with the color \033[1m{:}\033[0m.
    
Alternative hypothesis:
    There is a significant difference between the average revenue from products
        with the color \033[1m{:}\033[0m and products with the color \033[1m{:}\033[0m.\n
    """.format(c1,c2,c1,c2))
    print('Avg. revenue from products with {:} color:\t{:,.2f}$'.format(c1, sample_1.mean()))
    print('Avg. revenue from products with {:} color:\t{:,.2f}$\n'.format(c2, sample_2.mean()))
    print('alpha:',alpha)
    print('p-value:',results.pvalue,'\n')
    print('>> There is a {:.1%} probability of randomly getting a difference that size or larger.\n'.format(results.pvalue))
    if results.pvalue < alpha:
        print("""\033[1;42m We reject the null hypothesis, \033[0m
    there is a significant difference between the average revenue.\n""")
    else:
        print("\033[1;41m We can't reject the null hypothesis. \033[0m\n")
    print('{:->30}'.format(''))
```

```python
top_colors.remove('red')
for i in top_colors:
    rev_ttest('red',i)
```

### Summary

The red color leads with an average profit of almost 4.7K USD. If it was not enough to see the gaps on the graph, I ran a T test to test the significance of the differences between the averages of the red color and the other colors. Apart from the white, brown and yellow colors, the difference is significant enough to be included on all store products.
